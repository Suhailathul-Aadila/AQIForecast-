```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# Load all cities (do NOT filter for a specific one)
df = pd.read_csv("C:/Users/suhai/AQI_forecast/AQI_data/city_day/city_day.csv")  # Your actual file path
df = df[['Date', 'City', 'AQI']]
df.dropna(inplace=True)

# One-hot encode City
encoder = OneHotEncoder(sparse_output=False)
city_encoded = encoder.fit_transform(df[['City']])

# Normalize AQI
scaler = MinMaxScaler()
df['AQI_scaled'] = scaler.fit_transform(df[['AQI']])

# Combine features: [AQI, City One-Hot]
features = np.hstack([df['AQI_scaled'].values.reshape(-1, 1), city_encoded])


# import pandas as pd
# import numpy as np
# from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# # Load all cities (do NOT filter for a specific one)
# df = pd.read_csv("C:/Users/suhai/AQI_forecast/AQI_data/city_day/city_day.csv")  # Your actual file path
# df = df[[ 'City', 'AQI']]
# df.dropna(inplace=True)

# # One-hot encode City
# encoder = OneHotEncoder(sparse_output=False)
# city_encoded = encoder.fit_transform(df[['City']])

# # Normalize AQI
# scaler = MinMaxScaler()
# df['AQI_scaled'] = scaler.fit_transform(df[['AQI']])

# # Combine features: [AQI, City One-Hot]
# features = np.hstack([df['AQI_scaled'].values.reshape(-1, 1), city_encoded])
```

Create sequences with city tracking

```python
time_steps = 30

def create_sequences(data, labels, city_names, time_steps=30):
    X, y, cities = [], [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i + time_steps])
        y.append(labels[i + time_steps])
        cities.append(city_names[i + time_steps])
    return np.array(X), np.array(y), np.array(cities)

X, y, cities = create_sequences(
    data=features,
    labels=df['AQI_scaled'].values,
    city_names=df['City'].values,
    time_steps=time_steps
)
```

 Build and Train Functional CNN-LSTM Model

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, LSTM, Dense

# Define input shape: (time_steps, number of features)
input_shape = (X.shape[1], X.shape[2])  # e.g., (30, 1 + number_of_cities)

# Functional model definition
input_layer = Input(shape=input_shape)
x = Conv1D(filters=32, kernel_size=3, activation='relu')(input_layer)
lstm_out = LSTM(64, activation='relu', name="lstm_layer")(x)
output_layer = Dense(1)(lstm_out)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X, y, epochs=20, batch_size=32, validation_split=0.1)
```

 Extract LSTM features

```python
# Run model to extract features
intermediate_model = Model(inputs=model.input, outputs=model.get_layer('lstm_layer').output)
lstm_outputs = intermediate_model.predict(X)
```

Group features by city

```python
# Create DataFrame of LSTM outputs with associated city
lstm_df = pd.DataFrame(lstm_outputs)
lstm_df['City'] = cities

# Average LSTM features per city
city_embeddings = lstm_df.groupby('City').mean()
```

 Perform clustering

```python
from sklearn.cluster import KMeans

n_clusters = min(5, len(city_embeddings))  # Adjust cluster count based on number of cities
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
city_embeddings['Cluster'] = kmeans.fit_predict(city_embeddings)
```

Visualize clusters

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

pca = PCA(n_components=2)
pca_result = pca.fit_transform(city_embeddings.drop(columns='Cluster'))

plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=city_embeddings['Cluster'], s=100, palette='Set2')

# Label cities
for i, city in enumerate(city_embeddings.index):
    plt.text(pca_result[i, 0], pca_result[i, 1], city, fontsize=9)

plt.title("City Clusters Based on LSTM AQI Patterns")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```
