```python
import pandas as pd
import numpy as np

df=pd.read_csv('filtered_data.csv')
df=df.drop(["AQI_Bucket","City","Date"],axis=1)

print(df.isnull().sum())
print(df.duplicated().sum())
df=df.drop_duplicates()
print(df.duplicated().sum())
df.replace([np.inf,-np.inf],np.nan,inplace=True)
df.fillna(0,inplace=True)
```

```python
# pip install scikit-learn
```

```python
print('Null Values:',df.isnull().values.sum())

print('NA values:',df.isnull().values.any())

df.drop(df[df.AQI ==0].index, inplace=True)

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

import operator

y=df['AQI']
x=df.drop(['AQI'],axis=1)
fe=SelectKBest(score_func=chi2,k="all")
fe.fit(x,y)
colnm=[]
for col in x.columns:
    colnm.append(col)

fel=[]
for i in range(len(fe.scores_)):
    fel.append(fe.scores_[i])

dictt=dict(zip(colnm,fel))
cl_dictt={j:dictt[j]for j in dictt if not pd.isna(dictt[j])}
sort=dict(sorted(cl_dictt.items(),key=operator.itemgetter(1),reverse=True))
print(sort)
```

```python
import matplotlib.pyplot as plt
```

```python
names = list(sort.keys())
values = list(sort.values())
name=names
price=values

fig, ax=plt.subplots(figsize = (16,9))

ax.barh(name,price)

for s in ['top','bottom','left','right']:
    ax.spines[s].set_visible(False)

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

ax.xaxis.set_tick_params(pad=5)
ax.yaxis.set_tick_params(pad=10)

ax.grid(visible= True, color = 'grey',
        linestyle = '-.',linewidth = 0.5,
        alpha = 0.2)
ax.invert_yaxis()

for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
                str(round((i.get_width()),2)),
                fontsize = 10, fontweight = 'bold',
                color = 'grey')

ax.set_title('Feature Scores',
             loc = 'left')

plt.show()
```
